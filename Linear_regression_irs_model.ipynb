```python
import pandas as pd
import numpy as np
df = pd.read_csv("data.csv")
df.head()


```

```python
X=df[['ne_500','Te_500','ne_1000','Te_1000','ne_1500','Te1500','ne_2000','Te_2000','TEC']]
y=df['ssn']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
```

```python

nan_indices = np.where(df['TEC'].isna())[0]
print(nan_indices)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test,y_test)
```
